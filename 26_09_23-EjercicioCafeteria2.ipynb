```python
# Definir los tipos de comida y bebidas
comidas = ["Pizza", "Hamburguesa", "Sándwich", "Ensalada", "Pasta"]
bebidas = ["Café", "Té", "Refresco"]

# Inicializar lista de pedidos
pedidos = []

# Función para capturar alumnos de manera recursiva
def capturar_alumno():
    # Solicitar nombre del alumno
    nombre = input("Ingrese el nombre del alumno (o 'q' para salir): ")

    if nombre != 'q':
        pedidos_alumno = {}

        # Mostrar tipos de comida disponibles
        print("Tipos de comida disponibles:")
        for i, comida in enumerate(comidas, start=1):
            print(f"{i}. {comida}")
        comida_elegida = int(input("Elija el número de la comida que desea: "))
        pedidos_alumno["Comida"] = comidas[comida_elegida - 1]

        # Mostrar tipos de bebida disponibles
        print("Tipos de bebida disponibles:")
        for i, bebida in enumerate(bebidas, start=1):
            print(f"{i}. {bebida}")
        bebida_elegida = int(input("Elija el número de la bebida que desea: "))
        pedidos_alumno["Bebida"] = bebidas[bebida_elegida - 1]

        # Agregar el pedido a la lista de pedidos
        pedidos.append((nombre, pedidos_alumno))

        # Llamada recursiva para capturar más alumnos
        capturar_alumno()

        # Ordenar la lista de pedidos por nombre
    ordenar_pedidos = sorted(pedidos, key=lambda x: x[0])

    print("\nLista de nombres de alumnos ordenados:")
    for i, (nombre, _) in enumerate(ordenar_pedidos, start=1):
        print(f"{i}. Nombre: {nombre}")

# Función para generar pedidos (agregada)
def generar_pedidos():
    pass

# Función para imprimir resultados (agregada)
def imprimir_resultados():
    for i, (nombre, pedido) in enumerate(pedidos, start=1):
        print(f"{i}. Nombre: {nombre}, Comida: {pedido['Comida']}, Bebida: {pedido['Bebida']}")

# Función para ordenar pedidos por nombre (agregada)
def ordenar_pedidos():
    global pedidos
    pedidos = sorted(pedidos, key=lambda x: x[0])

# Función para agregar un pedido a la lista de pedidos (agregada)
def agregar_pedido(nombre, comida, bebida):
    pedidos.append((nombre, {"Comida": comida, "Bebida": bebida}))

# Función para ver la lista de pedidos (agregada)
def ver_pedidos():
    for i, (nombre, pedido) in enumerate(pedidos, start=1):
        print(f"{i}. Nombre: {nombre}, Comida: {pedido['Comida']}, Bebida: {pedido['Bebida']}")

# Función para eliminar un pedido de la lista (agregada)
def eliminar_pedido(numero_pedido):
    global pedidos
    if 1 <= numero_pedido <= len(pedidos):
        pedidos.pop(numero_pedido - 1)

# Función para aplicar el método sort a los pedidos (agregada)
def sort_pedidos():
    global pedidos
    pedidos = sorted(pedidos, key=lambda x: x[0])

# Función para usar el método pop de una pila (agregada)
def pop_pila():
    if len(pedidos) > 0:
        return pedidos.pop()
    else:
        print("La lista de pedidos está vacía.")

# Función para usar el método push en una pila (agregada)
def push_pila(nombre, comida, bebida):
    pedidos.append((nombre, {"Comida": comida, "Bebida": bebida}))

# Función para usar el método insert en la lista de pedidos (agregada)
def insert_pedido(posicion, nombre, comida, bebida):
    pedidos.insert(posicion, (nombre, {"Comida": comida, "Bebida": bebida}))

# Función para contar la cantidad de pedidos (agregada)
def contar_pedidos():
    return len(pedidos)

# Menú de opciones
while True:
    print("\n--- Menú ---")
    print("1. Capturar Alumno")
    print("2. Generar Pedidos")
    print("3. Imprimir Resultados")
    print("4. Ordenar Pedidos por Nombre")
    print("5. Ver Pedidos")
    print("6. Eliminar Pedido")
    print("7. Aplicar sort a los Pedidos")
    print("8. Pop de una Pila")
    print("9. Push en una Pila")
    print("10. Insertar Pedido en Posición Específica")
    print("11. Contar Pedidos")
    print("12. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == '1':
        capturar_alumno()
    elif opcion == '2':
        generar_pedidos()
    elif opcion == '3':
        imprimir_resultados()
    elif opcion == '4':
        ordenar_pedidos()
    elif opcion == '5':
        ver_pedidos()
    elif opcion == '6':
        numero_pedido = int(input("Ingrese el número de pedido que desea eliminar: "))
        eliminar_pedido(numero_pedido)
    elif opcion == '7':
        sort_pedidos()
    elif opcion == '8':
        pop_pila()
    elif opcion == '9':
        nombre = input("Ingrese el nombre del alumno: ")
        comida = input("Ingrese la comida: ")
        bebida = input("Ingrese la bebida: ")
        push_pila(nombre, comida, bebida)
    elif opcion == '10':
        posicion = int(input("Ingrese la posición donde desea insertar el pedido: "))
        nombre = input("Ingrese el nombre del alumno: ")
        comida = input("Ingrese la comida: ")
        bebida = input("Ingrese la bebida: ")
        insert_pedido(posicion, nombre, comida, bebida)
    elif opcion == '11':
        cantidad_pedidos = contar_pedidos()
        print(f"Hay {cantidad_pedidos} pedidos en la lista.")
    elif opcion == '12':
        break
    else:
        print("Opción no válida. Intente de nuevo.")

```
