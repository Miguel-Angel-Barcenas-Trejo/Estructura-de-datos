```python
def jugar(intento=1):
    respuesta = input("¿De qué color es una naranja? ")
    if respuesta != "naranja":
        if intento < 3:
            print("\nFallaste, inténtelo de nuevo")
            intento += 1
            jugar(intento)  # llamada recursiva
        else:
            print("\nPerdiste.")
    else:
        print("\n¡Ganaste!")

jugar()
```
