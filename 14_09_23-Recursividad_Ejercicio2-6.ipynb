```python
def factorial(fact):
  if fact >0:
    valor = fact * factorial(fact-1)
    return valor
  else:
      return 1;

resultado = factorial(0)
print(f"El factorial de  es {resultado}")

```

```python
def ordenar(lista, cant):
  if cant >1:
    for f in range(0, cant-1):
      if lista[f] > lista[f+1]:
        aux = lista[f]
        lista[f] = lista[f + 1]
        lista[f + 1] = aux
        ordenar(lista, cant -1)

datos = [60,44,22,33,21]
print(datos)
ordenar(datos, len(datos))
print(datos)
```
